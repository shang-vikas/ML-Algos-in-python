```python
import numpy as np
import time
import sys
LARGE_NUMBER = sys.maxsize
```

```python
class Dataset:
    def __init__(self,X,Y):
        self.X = X
        self.Y = Y
```

```python
class TreeNode:
    '''
    Class for a node of a tree..
    JOB - calculate the gain and decide whether to split the node.
    '''
    def __init__(self):
        self.is_leaf = False
        self.left_child = None
        self.right_child = None
        self.split_feature_id = None
        self.split_val = 0
        self.weight = 0
        
        
    def _calc_split_gain(self,grad,hess,G_l,H_l,G_r,H_r,lambda_):
        
        def calc_term(g,h):
            return np.square(g)/(h+lambda_)
        return calc_term(G_l,H_l) + calc_term(G_r,H_r) - calc_term(grad,hess)
    
    def _calc_leaf_weight(self,grad,hess,lambda_):
        return np.sum(grad)/(np.sum(grad)+lambda_)
    
    def build(self,instances,grad,hess,shrinkage,param,depth):
        assert instances.shape[0] == grad.__len__() == hess.__len__()
        
        if depth > param['max_depth']:
            self.is_leaf = True
            self.weight = self._calc_leaf_weight(grad,hess,param['lambda'])*shrinkage
            return
        G = np.sum(grad)
        H = np.sum(hess)
        best_gain = 0.
        best_feature_id = None
        best_val = 0.
        best_left_instances = None
        best_right_instances = None
        for feature_id in range(instances.shape[1]):
            G_l = 0.
            H_l = 0.
            sorted_instance_ids = instances[:,feature_id].argsort()
            for idxx in range(sorted_instance_ids.shape[0]):
                G_l += grad[sorted_instance_ids[idxx]]
                H_l += hess[sorted_instance_ids[idxx]]
                G_r = G - G_l
                H_r = H - H_l
                current_gain = _calc_split_gain(G,H,G_l,H_l,G_r,H_r,param['lambda'])
                if current_gain > best_gain:
                    best_gain = current_gain
                    best_feature_id = feature_id
                    best_left_instances = sorted_instance_ids[:idxx+1]
                    best_right_instances = sorted_instance_ids[idxx+1:]
                    best_val = instances[sorted_instances_ids[idxx]][feature_id]
            if best_gain < param['min_split_gain']:
                self.is_leaf = True
                self.weight = self._calc_leaf_weight(grad,hess,param['lambda'])*shrinkage
                
            else:
                self.split_feature_id = best_feature_id
                self.split_val = best_val
                self.left_child = TreeNode()
                self.left_child.build(best_left_instances,G_l,H_l,shrinkage,param,depth+1)
                self.right_child = TreeNode()
                self.right_child.build(best_right_instances,G_r,H_r,shrinkage,param,depth+1)
                
                
    def predict(self,x):
        if self.is_leaf:
            return self.weight
        else:
            ## come again here and understand the code
            if x[self.split_feature_id] >=self.split_val:
                self.left_child.predict(x)
            else:
                self.right_child.predict(x)
    
    

class Tree(object):
    
    def __init__(self):
        self.root = None
        
    def build(self,instances,grad,hess,shrinkage,param):
        assert instances.shape[0] == len(grad) == len(hess)
        self.root = TreeNode()
        current_depth = 0
        self.root.build(instances,grad,hess,shrinkage,param,current_depth)
        
    def predict(self,instances):
        return self.root.predict(instances)
    
    
class GBT(object):
    
    def __init__(self,num_estimators=100,objective='reg',lambd=1,gamma=0,
                 min_split_gain=0.1,max_depth=5,lr=0.03):
        self.params = {
            'num_estimators':num_estimators,
            'objective':objective,
            'lambda':lambd,
            'gamma':gamma,
            'min_split_gain':min_split_gain,
            'max_depth':max_depth,
            'shrinkage':lr
            
        }
        
        self.best_iteration = None
        
    def _calc_training_data_scores(self,train_set,models):
        if len(models)==0:
            return None
        else:
            X = train_set.X
            scores = np.zeros(len(X))
            for idxx in range(len(X)):
                scores[idxx] = self.predict(X[idxx],models=models)
            
            return scores
            
            
    def _calc_l2_gradient(self train_set, scores):
        labels = train_set.Y
        hessian = np.full(len(labels),2)
        if scores is None:
            grad = np.random.uniform(size=len(labels))
        else:
            grad = np.array([2*(labels[i] - scores[i]) for i in range(len(labels))])
        
        return grad,hessian
    
    
    def _calc_gradient(self,train_set,scores):
        return self._calc_l2_gradient(train_set,scores)
    
    
    def _calc_l2_loss(self,train_set,models):
        # errors = []
        score = np.array([(train_set.Y[i] - self.predict(train_set.X[i]),models) for i in range(len(train_set))])
        
        return np.mean(np.square(score))
    
    
    def _calc_loss(self,train_set,models):
        return self._calc_l2_loss(train_set,models)
            
    
    def _build_learner(self,train_set,grad,hessian,shrinkage):
        learner = Tree()
        learner.build(train_set.X,grad,hessian,shrinkage,self.params)
        return learner
    
    
    def train(self,num_boost_rounds=100,params,train_set,valid_set=None,early_stopping_rounds=10):
        self.params.update(params)
        models = []
        shrinkage = 1.
        best_iteration = None
        best_val_loss = LARGE_NUMBER
        train_start_time = time.time()
        if valid_set is None:
            print('Validation set is not Provided')
        for count in num_boost_round:
            iter_start_time = time.time()
            
            
            
        
        
```

```python
from numba import jit
from numpy import arange
```

```python

# jit decorator tells Numba to compile this function.
# The argument types will be inferred by Numba when function is called.
def _sum2d(arr):
    M, N = arr.shape
    result = 0.0
    for i in range(M):
        for j in range(N):
            result += arr[i,j]
    return result

@jit
def sum2d(arr):
    M, N = arr.shape
    result = 0.0
    for i in range(M):
        for j in range(N):
            result += arr[i,j]
    return result
```

```python
import sklearn.datasets as datasets
import pandas as pd
iris=datasets.load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
y=iris.target

from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(max_depth=3)
dtree.fit(df,y)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()

export_graphviz(dtree, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
